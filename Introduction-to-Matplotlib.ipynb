## Matplotlin example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots( figsize = (7, 7)) #width, height

#3. Plot data
ax.plot(x, y) 

# 4. Customize plot
ax.set(title = "Simple Plot",
       xlabel="x-axis",
       ylabel = "y-axis")

# 5. Save your plot and show
fig.savefig("images/sample-plot.png")
```

## Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
# The issue here is that plt.subplots() returns a tuple containing a figure 
# and an array of subplots (axes), not just the figure itself.
# When you use ax, = plt.subplots(), you are unpacking the tuple 
# and assigning the first element (the figure) to the variable ax. 
# However, since you're only working with one subplot
# you should use fig, ax = plt.subplots() without the comma after ax 
# to correctly unpack the tuple.
```

```python
# Use same data to make a scatter 
x = np.linspace(0, 10, 100)
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Use same data to make a scatter 
x = np.linspace(0, 10, 100)
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
#Make a plot from dictionary
butter_prices = {
    "Almond butter" : 10,
    "Peanut butter" : 8,
    "Cashew butter" : 12,
}

fig, ax = plt.subplots()
ax.bar(butter_prices.keys(), butter_prices.values())
ax.set(title="Nut Butter Store",
      ylabel = "Prices ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()))
```

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows = 2,
                                            ncols= 2,
                                            figsize=(10,5))
ax1.bar(butter_prices.keys(), butter_prices.values());
ax2.scatter(x, np.sin(x))
ax3.plot(x, x/2)
ax4.hist(np.random.rand(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize = (10,5))

#plot to each different index
ax[0, 0].bar(butter_prices.keys(), butter_prices.values());
ax[0, 1].scatter(x, np.sin(x))
ax[1,0].plot(x, x/2)
ax[1, 1].hist(np.random.rand(1000));
```

```python
car_sales["Price"] = car_sales["Price"].replace('[\$\,\.]', '', regex=True)
car_sales
```

```python
# Remove last two zeros and convert to integers
car_sales["Price"] = car_sales["Price"].str[:-2]

car_sales
```

```python
car_sales["Sale Date"] = index=pd.date_range("1/1/2020", periods=10)
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales["Price"]
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
# Plot scatter plot with prce column as numeric
car_sales.plot(x="Odometer (KM)", y = "Price", kind = "scatter")
```
