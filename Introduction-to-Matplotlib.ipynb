## Matplotlin example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots( figsize = (7, 7)) #width, height

#3. Plot data
ax.plot(x, y) 

# 4. Customize plot
ax.set(title = "Simple Plot",
       xlabel="x-axis",
       ylabel = "y-axis")

# 5. Save your plot and show
fig.savefig("images/sample-plot.png")
```

## Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
# The issue here is that plt.subplots() returns a tuple containing a figure 
# and an array of subplots (axes), not just the figure itself.
# When you use ax, = plt.subplots(), you are unpacking the tuple 
# and assigning the first element (the figure) to the variable ax. 
# However, since you're only working with one subplot
# you should use fig, ax = plt.subplots() without the comma after ax 
# to correctly unpack the tuple.
```

```python
# Use same data to make a scatter 
x = np.linspace(0, 10, 100)
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Use same data to make a scatter 
x = np.linspace(0, 10, 100)
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
#Make a plot from dictionary
butter_prices = {
    "Almond butter" : 10,
    "Peanute butter" : 8,
    "Cashew butter" : 12,
}

fig, ax = plt.subplots()
ax.bar(butter_prices.keys(), butter_prices.values())
ax.set(title="Nut Butter Store",
      ylabel = "Prices ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()))
```
